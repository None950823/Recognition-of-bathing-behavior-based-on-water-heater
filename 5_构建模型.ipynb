```python
train_feature = ['用水事件总时长(Min)','用水时长(Min)','用水时长/总用水时长(Min)','总停顿时长','平均停顿时长','停顿次数','总用水量',
                '平均水流量','水流量波动','停顿时长波动']
train_data = data[train_feature].values
label = data['是否为洗浴（1表示是,0表示否）'].values
X_train, X_test, y_train, y_test = train_test_split(train_data, label, test_size=0.3, random_state=42)
X_train = preprocessing.StandardScaler().fit_transform(X_train)
X_test = preprocessing.StandardScaler().fit_transform(X_test) 
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report
# 建立模型
bpnn = MLPClassifier(hidden_layer_sizes=(15, 12),
                     max_iter=200,
                     solver='lbfgs',
                     random_state=50)
bpnn.fit(X_train, y_train)
y_pred = bpnn.predict(X_test)  # 返回预测结果
print('神经网络预测结果评价报告：\n', classification_report(y_test, y_pred))
```

```python
#建立、训练多层神经网络，并完成模型的检验

from keras.models import Sequential
from keras.layers import Dense
# 声明序贯模型
model=Sequential()
model.add(Dense(units=17,
                input_dim=10,
                kernel_initializer='normal',
                activation='relu'))
model.add(Dense(units=10,
                kernel_initializer='normal',
                activation='relu'))
model.add(Dense(units=1,
                kernel_initializer='normal',
                activation='sigmoid'))
model.compile(loss='binary_crossentropy',
              optimizer='adam',metrics=['accuracy'])
train_history=model.fit(x=X_train,
                        y=y_train,
                        validation_split=0.2,
                        epochs=10,
                        batch_size=200,
                        verbose=2)
```
