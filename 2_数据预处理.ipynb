# 2、数据变换

## 从数据中识别出哪些记录是一个完整的用水事件（包括洗脸、洗手、洗浴等），因该项目是对洗浴事件的识别，所以为了提高在大量的一次完整用水事件的效率，需要建立筛选规则剔除掉那些明显不是洗浴事件的记录。

## 怎样划分一个完整的用水事件？根据水流量和停顿时间间隔的阈值去划分，所以下面建立阈值寻优模型

## 2.1 一次完整用水事件的划分模型
### 在数据集中，记录了各种用水事件，包括 洗浴，洗手，刷牙，洗脸，洗菜等，而且一次用水事件由数条甚至数千条的状态记录组成，所以首先需要在大量的状态记录中划分出那些连续的数据是一次完整的用水事件

### 划分方法的大致思路：水流量不为0表明用户正在使用热水；而水流量为0表明用户用热水发生停顿或者用热水结束。如果水流量为0的记录之间的时间间隔超过一个阈值T，则从该段水流量为0的状态记录向前找到最后一条水流量不为0的用户用水记录作为上一次用水事件的结束；向后找到水流量不为0的状态记录作为下一个用水事件的开始。
### 一次完整用水事件的划分步骤如下：
### 1）读取数据记录，识别到第一条水流量不为0的数据记录记为R1，按顺序识别接下来的一条水流量不为0数据记录为R2，记录其之间的时间间隔gap = R2 - R1
### 2) 若gap > T（时间间隔阈值），则此次R1和R2及之间的数据记录不能划分到同一次用水事件中。同时将R2记录作为新的读取数据记录的开始，返回步骤1）；若gap< T，则将R2和R1之间数据记录划分到同一次用水事件中。
### 3）循环执行步骤2），直到数据记录读取完毕， 结束事件划分

```python
threshold = pd.Timedelta(minutes = 4)
heater_data['发生时间'] = pd.to_datetime(heater_data['发生时间'],format = '%Y%m%d%H%M%S')
data  = heater_data[heater_data['水流量'] > 0] #只要水流量大于0的记录
d = data['发生时间'].diff() > threshold #相邻时间做差，比较是否大于阈值
data['事件编号'] = d.cumsum() + 1 #通过累加的求和的方式为事件编号
data.to_csv('./dividsequence.csv',index = False)
data
```

## 2.2 用水事件阈值寻优模型

### 受时间、地域、季节等因素的影响，使用固定的阈值显然是不合理的
### 需要建立一个阈值寻优模型来更新寻找最优的阈值

```python
#先统计不同阈值下，划分事件个数的区别
threshold = np.arange(1,10.25,0.25)
num_with_differentThr = []
for thr in threshold:
    t = pd.Timedelta(minutes = thr)
    d = data['发生时间'].diff() > t #相邻时间做差，比较是否大于阈值
    num_with_differentThr.append((d.cumsum() + 1).nunique())
```

```python
eventNum_with_different_thr = pd.DataFrame(columns = ['时间间隔'])
eventNum_with_different_thr['时间间隔'] = threshold
eventNum_with_different_thr['事件个数'] = num_with_differentThr
eventNum_with_different_thr
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei'] # 步骤一（替换sans-serif字体）
plt.rcParams['axes.unicode_minus'] = False 
plt.figure(figsize=(10,5))#设置画布的尺寸
plt.title('阈值与划分事件个数关系',fontsize=20)#标题，并设定字号大小
plt.xlabel(u'时间间隔',fontsize=14)#设置x轴，并设定字号大小
plt.ylabel(u'事件数',fontsize=14)#设置y轴，并设定字号大小
plt.plot(eventNum_with_different_thr['时间间隔'],eventNum_with_different_thr['事件个数'],color="deeppink",linewidth=2,linestyle=':',label='Jay income', marker='o')

plt.show()
```

### 上图总结：事件个数是衡量用水频繁度的。 在上图曲线中可以看出，在阈值较小和较大两头部分，下降趋势明显，说明在该段阈值范围内，用户的听读习惯比较集中；如果想曲线中间段那样，趋势比较平缓，则说名用户的停顿习惯趋于稳定，所以去该段时间开始作为阈值，既不会将短的用水事件合并，又不会将长的用水事件拆开

# 阈值寻优思路

## 根本思想：找到斜率最小的点对应的时间间隔
## 假设当前时间间隔为3，它对应的事件次数186，在3之后还有很多时间内间隔，为了计算当前时间间隔3所对应的斜率，只取3之后的四个点（算上本身3就是五个点），分别标记为A(3),B(3.25),C(3.5),D(3.75),E(4)，其对应的事件个数分别为186、181、178、174、172。那么使用如下公式计算当前点A对应的斜率值：
$$K =\frac{k_{AB} + k_{BC}+ k_{CD} + k_{DE}}{4}$$
## 其中$k_{ij}=|\frac{y_{i} - y_{j}}{x_{i} - x_{j}}|$

### 举例：A(3,186),B(3.25,181),C(3.5,178),D(3.75,174),E(4,173)
$$k_{AB} = |\frac{186 - 181}{3 - 3.25}|=24$$
$$k_{BC} = |\frac{181 - 178}{3.25 - 3.5}|=12$$
$$k_{CD} = |\frac{178 - 174}{3.5 - 3.75}|=16$$
$$k_{DE} = |\frac{174 - 173}{3.75 - 4}|=4$$
### 那么 $$K_{A} =\frac{k_{AB} + k_{BC}+ k_{CD} + k_{DE}}{4}= \frac{24 +12 + 16+4}{4}=14$$
### 既当前点对应的斜率就是14，其他点以此类推。
## 这样做会有一个问题哈，就是最后的4个点是不能够计算这个斜率值得(因为点不够)，为了解决这个问题，可以设置一个阈值上限 ，使最后的4个阈值不是考虑范围内的阈值。

## 根据上面的计算， 会有如下两种情况：

### 1) 当存在一个阈值的斜率指标K <1时，则取阈值最小的点A（为什么去最小的？因为可能存在多个阈值的斜率指标小于1）的横坐标$x_{A}$作为用水事件划分的阈值（取1是因为专家阈值）

### 2) 当不存在K<1时，则查找所有阈值中斜率指标最小的阈值。如果该阈值的斜率指标小于5，则取该阈值作为用水时间划分的阈值；如果该阈值 的斜率指标不小于5，则该阈值去默认值4。（4和5均为专家阈值）


```python
def event_num(ts):
    ts = pd.Timedelta(minutes = ts)
    d = data['发生时间'].diff() > ts  #相邻时间做差，比较是否大于阈值
    return (d.cumsum() + 1).nunique()                #这样直接返回事件数

h = pd.DataFrame(threshold,columns = ['阈值']) #定义阈值列
h['事件数'] = h['阈值'].apply(event_num)
h['斜率'] = h['事件数'].diff(-1) / 0.25  #计算每两个相邻点对应的斜率
h['斜率'].fillna(0,inplace = True)
tmp= [] 
for i,j in enumerate(h['斜率'].abs()):
    if i <= (len(h['斜率']) - 4):
        tmp.append((h['斜率'].abs()[i]+h['斜率'].abs()[i+ 1]+ h['斜率'].abs()[i+2]+ h['斜率'].abs()[i+3] ) / 4)
    else:
        tmp.append(0)
h['斜率指标'] = tmp#采用后n个斜率绝对值平均作为斜率指标
h.loc[h['斜率指标'] == 0,'斜率指标'] = 1000    #为了防止最后3个对最小值求取的影响，将0用一个较大的数填充
ts = h['阈值'][h['斜率指标'].idxmin()] # 用idxmin返回最小值的Index，由于rolling_mean()自动计算的是前n个斜率的绝对值平均，所以结果要进行平移（-n）
if ts > 5:#5为专家阈值
    ts = pd.Timedelta(minutes = 4)
print('当前时间间隔划分方式下，最佳时间间隔是：',ts)
h
```
