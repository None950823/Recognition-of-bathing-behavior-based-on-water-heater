# 制作标签：筛选“候选洗浴事件”
## 在划分好的一次用水事件中识别出哪些一次用水事件是不是洗浴事件，那么剩下的就是洗浴事件

## 使用如下判断条件来标记是否为洗浴事件：
### 1）一次用水事件中总用水量(纯热水)小于Y升
### 2）用水时长小于100秒
### 3）总用水时长小于120秒

### 上面的判断条件有一个问题，就是需要确定一下总用水量的阈值Y，下面是判断方案：
### （1）洗澡的水温一般是37-41度，热水器设定的温度越高，热水器水的实际温度也越高，此时热水器的热水的使用量就越少；
### （2）实验分析，热水器设定温度为50度时，一次普通的洗浴时长为15分钟，总用水时长10分钟左右，热水的使用量在10-15升；
### （3）为了不影响特殊短暂的洗浴事件，以及考虑到下天用的热水较少，放宽范围假定热水器子设定温度为50度时，一次洗浴的总用水量为5升，同时取洗浴温度的均值为39度来计算热水器不同设定温度下的热水使用量阈值

## 标准热水量换算模型
### 洗浴用水温度：T（℃）
### 设定温度：X（℃）
### 自来水水温：C（℃）
### 自来水注入量： M（升）
### 设定温度为X时的用水量：Y（升）
### 50℃时的用水量：V（升）
### 假设两次洗浴事件热水和冷水混合后的花洒出水温度恒为T℃，总用水量不变且为 (M+V)，根据热量守恒建立方程组：
### $$\left\{\begin{matrix} (50 -T) * V + (C + T) *M = 0---- (1)\\(X - T) * Y + (C -T) * (M+V-Y)=0----(2)  \end{matrix}\right.$$
### (1)式是50℃的热水V升与M升C℃自来水混合得到M+V升T℃的洗浴用水的热守恒公式
### (2)式是X℃的热水Y升与M+V-Y升C℃自来水混合得到M+V升T℃的洗浴用水的热守恒公式

### 从而得到Y与X、C、V之间的关系：
### $$Y = \frac{(50-C)* V}{X -C}$$

#### 其中，V是热水器的水恒为50℃时洗浴时的最低用水量

### 根据上式可以计算出用水事件在不同实际用水温度下的标准热水使用量。其中的自来水每月平均温度取平均室温23℃

```python
data['是否为洗浴（1表示是,0表示否）'] = 2
```

```python
data.loc[data['用水时长(Min)'] < ((100) / 60),'是否为洗浴（1表示是,0表示否）' ] = 0
data.loc[data['用水事件总时长(Min)'] < ((120) / 60),'是否为洗浴（1表示是,0表示否）' ] = 0
data.loc[data['总用水量'] < 5,'是否为洗浴（1表示是,0表示否）' ] = 0
data.loc[data['是否为洗浴（1表示是,0表示否）'] == 2,'是否为洗浴（1表示是,0表示否）' ] = 1
```

```python
# 绘制条形图
fig = plt.figure(figsize=(6, 5))  # 设置画布大小
sns.set(style="darkgrid")
sns.set_context("notebook", font_scale=1.2, rc={"lines.linewidth": 2.5})
plt.rcParams['font.sans-serif'] = 'SimHei'  # 设置中文显示
plt.rcParams['axes.unicode_minus'] = False
sns.catplot(kind="count",
            data=data,
            x="是否为洗浴（1表示是,0表示否）",
            aspect=1,
            height=5,
            palette='tab20')
plt.xlabel('是否为洗浴')
plt.ylabel('比例')
plt.title('洗浴事件占比')
```
